```python
import pandas as pd
import json
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
with open("../../data/original/wine-info.json") as f:
  wine_info = json.load(f)
```

```python
wine_info_scraped = pd.DataFrame(wine_info)
wine_info_scraped
```

```python
food_list = []
for pairings in [f for f in wine_info_scraped["food_pairings"]]:
  food_list.extend(pairings)
food_list = list(set(food_list))
food_list
```

```python
# with open("../../data/food_list.json", "w") as f:
#   json.dump(food_list, f)
```

```python
# One hot encode wine_style
df = pd.get_dummies(wine_info_scraped, prefix="wine_style", columns=["wine_style"])

mlb = MultiLabelBinarizer()
df = wine_info_scraped.join(pd.DataFrame(mlb.fit_transform(wine_info_scraped['food_pairings']),columns=mlb.classes_).add_prefix("food_"))
df.drop(columns=["wine_style", "food_pairings"], inplace=True)
df
```

```python
df.to_csv("../../data/wine_info.csv", index=False)
```
